```python
def gen_commands2(host, group, config, linesep='\n'):
    commands = []
    c = config[host]
    is_avx2 = 'avx2' in c
    for best_coverage_type in best_coverage_types:
        command = get_command_line2(
            c['hash_func'],
            best_coverage_type,
            c['keys_base_dir'],
            group,
            c['output_base_dir'],
            c['concurrency'],
            is_avx2,
        )
        if 'no_timemem' in c:
            command = command.replace('timemem ', '')
        commands.append(command)
    return linesep.join(commands)  
```
