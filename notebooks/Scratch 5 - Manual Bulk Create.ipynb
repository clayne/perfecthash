```python
import os
import threading
import subprocess

import psutil

import numpy as np

from tpn.path import (
    join_path,
)

from tpn.clipboard import cb
```

```python
root = join_path(os.getcwd(), '..')
create_exe = join_path(root, 'src\\x64\\Release\\PerfectHashCreate.exe')
```

```python
keys_root_sys32 = 'c:\\src\\phkeys\\sys32'
output_path = 'c:\\temp\\phkeys.out.sys32'
```

```python
keys_paths = [ join_path(keys_path, f) for f in os.listdir(keys_root_sys32) ]
```

```python
#psutil.cpu_times_percent(interval=0, percpu=False)
```

```python
class PerfectHashCreateExe:
    def __init__(self, *args, **kwds):
        self.output_path = output_path
        self.algorithm = 'Chm01'
        self.hash_function = 'ShiftMultiplyXorShift'
        self.mask_function = 'And'
        self.concurrency = '1'
        self.extra = [
            '--AttemptsBeforeTableResize=10',
            '--MaxNumberOfTableResizes=0',
            '--IgnorePreviousTableSize',
            '--NoFileIo'
        ]
    
    def make_commandline_string(self, keys_path):
        s = '%s %s %s %s %s %s ' % (
            keys_path,
            self.output_path,
            self.algorithm,
            self.hash_function,
            self.mask_function,
            self.concurrency,
        )
        s += ' '.join(self.extra)
        return s
    
    def make_commandline(self, keys_path):
        l = [
            create_exe,
            keys_path,
            self.output_path,
            self.algorithm,
            self.hash_function,
            self.mask_function,
            self.concurrency,
        ]
        l += self.extra
        return l

    def run(self, keys_path):
        args = self.make_commandline(keys_path)
        before = psutil.cpu_times_percent(interval=0, percpu=False)
        proc = subprocess.Popen(args)
        after = psutil.cpu_times_percent(interval=0, percpu=False)
        return (proc, before, after)
```

```python
creator = PerfectHashCreateExe()
```

```python
def get_random_paths(keys_paths, count):
    length = len(keys_paths)
    a = np.unique(np.random.randint(0, length, count * 4))[:count]
    assert len(a) == count, (len(a), count)
    return [ keys_paths[i] for i in a ]
```

```python
procs = []
times = []
paths = get_random_paths(keys_paths, 3000)
```

```python
for p in paths:
    #print(p)
    (proc, before, after) = creator.run(p)
    procs.append(proc)
    times.append((before, after))
    break
    
```
